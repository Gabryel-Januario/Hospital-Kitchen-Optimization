```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sqlalchemy import create_engine

connection_string = "postgresql+psycopg2://postgres:123@localhost:5432/airflow"
engine = create_engine(connection_string)

children_rejected_sql_query = "SELECT * FROM dbt_models.fct_children_rejected"
admissions_daily_sql_query = "SELECT * FROM dbt_models.fct_admissions_daily"
consumption_events_sql_query = "SELECT * FROM dbt_models.fct_consumption_events"
rejected_meals_sql_query = "SELECT * FROM dbt_models.fct_rejected_meals"
ward_consumption_sql_query = "SELECT * FROM dbt_models.fct_ward_consumption"


children_rejected_df = pd.read_sql_query(children_rejected_sql_query, engine)
admissions_daily_df = pd.read_sql_query(admissions_daily_sql_query, engine)
consumption_events_df = pd.read_sql_query(consumption_events_sql_query, engine)
rejected_meals_df = pd.read_sql_query(rejected_meals_sql_query, engine)
ward_consumption_df = pd.read_sql_query(ward_consumption_sql_query, engine)
```

```python
print(children_rejected_df.head())
```

```python
plt.figure(figsize=(6,8))
sns.barplot(data=children_rejected_df, x='age_group', y='rejected')
plt.xlabel('Faixa Etária', size=12)
plt.ylabel('Quantidade de Rejeições', size=12)
plt.title('Quantidade de Rejeições por Faixa Etária', size=16)
plt.tight_layout()
plt.show()
            
```

```python
print(admissions_daily_df.head())
```

```python
plt.figure(figsize=(10,8))
sns.lineplot(data=admissions_daily_df, x='ala_hospitalar', y='total_de_crianças_admitidas')
plt.title('Total de Crianças Admitidas por Ala Hospitalar', size=16)
plt.xlabel('Ala Hospitalar', size=12)
plt.ylabel('Total de Crianças Admitidas', size=12)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.lineplot(data=admissions_daily_df, x='faixa_etaria', y='total_de_crianças_admitidas')
plt.title('Total de Crianças Admitidas por Faixa Etária', size=16)
plt.xlabel('Faixa Etária', size=12)
plt.ylabel('Total de Crianças Admitidas', size=12)
plt.tight_layout()
plt.show()
```
