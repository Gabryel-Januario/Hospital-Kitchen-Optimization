# Análise Exploratória de Dados (EDA) - Cozinha Hospitalar

#### Objetivo

Este notebook tem como objetivo realizar uma análise exploratória dos dados de uma cozinha hospitalar pediátrica. Buscaremos entender os padrões de consumo, rejeição de refeições e desperdício de alimentos, cruzando informações sobre os pacientes (faixa etária, sexo) e as alas hospitalares.

### Estrutura da Análise

*Carregamento e Preparação dos Dados:* Conexão com o banco de dados e carregamento das tabelas.

*Análise Geral de Rejeições:* Foco nas rejeições por faixa etária.

*Análise de Admissões Hospitalares:* Perfil das crianças admitidas por ala, faixa etária e sexo.

*Análise de Refeições Rejeitadas:* Identificação dos pratos com maior índice de rejeição.

*Análise de Consumo e Desperdício:* Investigação sobre o desperdício por ala e tipo de refeição.

*Conclusão:* Resumo dos principais insights encontrados.

### 1. Carregamento e Preparação dos Dados

Primeiramente, vamos importar as bibliotecas necessárias e estabelecer a conexão com o banco de dados para carregar nossos conjuntos de dados.

```python
# Importação das bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine

# Configuração do estilo dos gráficos
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)
```

```python
# Conexão com o banco de dados PostgreSQL
connection_string = "postgresql+psycopg2://postgres:123@localhost:5432/airflow"
engine = create_engine(connection_string)

# Queries para carregar os dados
children_rejected_sql_query = "SELECT * FROM dbt_models.fct_children_rejected"
admissions_daily_sql_query = "SELECT * FROM dbt_models.fct_admissions_daily"
consumption_events_sql_query = "SELECT * FROM dbt_models.fct_consumption_events"
rejected_meals_sql_query = "SELECT * FROM dbt_models.fct_rejected_meals"
ward_consumption_sql_query = "SELECT * FROM dbt_models.fct_ward_consumption"

# Carregando os DataFrames
children_rejected_df = pd.read_sql_query(children_rejected_sql_query, engine)
admissions_daily_df = pd.read_sql_query(admissions_daily_sql_query, engine)
consumption_events_df = pd.read_sql_query(consumption_events_sql_query, engine)
rejected_meals_df = pd.read_sql_query(rejected_meals_sql_query, engine)
ward_consumption_df = pd.read_sql_query(ward_consumption_sql_query, engine)
```

### Visualização Rápida dos Dados

Vamos inspecionar as primeiras linhas de cada DataFrame para entender a estrutura dos dados.

```python
print("--- Crianças por Rejeição ---")
print(children_rejected_df.head())
print("\\n--- Admissões Diárias ---")
print(admissions_daily_df.head())
print("\\n--- Pratos Rejeitados ---")
print(rejected_meals_df.head())
print("\\n--- Consumo por Ala ---")
print(ward_consumption_df.head())
```

###  2. Análise Geral de Rejeições por Faixa Etária

Nesta seção, vamos analisar a quantidade total de refeições rejeitadas, segmentando por faixa etária das crianças. Isso pode nos dar um primeiro insight sobre qual grupo de idade é mais propenso a rejeitar as refeições oferecidas.

```python
# Gráfico de barras para rejeições por faixa etária
plt.figure(figsize=(8, 6))
sns.barplot(data=children_rejected_df, x='age_group', y='rejected', palette='viridis', order=children_rejected_df.sort_values('rejected', ascending=False).age_group)

plt.title('Quantidade Total de Rejeições por Faixa Etária', size=16, fontweight='bold')
plt.xlabel('Faixa Etária', size=12)
plt.ylabel('Quantidade de Rejeições', size=12)
plt.tight_layout()
plt.show()
```

#### *Análise do Gráfico*

- Crianças (Criança) representam o grupo com o maior número absoluto de rejeições.

- Adolescentes vêm em segundo lugar, com um número também significativo.

- Bebês (Bebê) têm o menor número de rejeições, o que é esperado, já que sua alimentação é, em geral, mais específica e controlada.

A alta rejeição no grupo "Criança" pode indicar uma oportunidade para rever os cardápios oferecidos a essa faixa etária.

### 3. Análise de Admissões Hospitalares

Agora, vamos explorar os dados de admissões para entender a distribuição das crianças pelas alas hospitalares e faixas etárias. Isso nos dará contexto sobre onde a maior parte da população infantil do hospital está concentrada.

```python
# Gráfico de barras para o total de crianças admitidas por ala
plt.figure(figsize=(12, 7))
sns.barplot(data=admissions_daily_df, x='ala_hospitalar', y='total_de_crianças_admitidas', estimator=sum, errorbar=None, palette='plasma')

plt.title('Total de Crianças Admitidas por Ala Hospitalar', size=16, fontweight='bold')
plt.xlabel('Ala Hospitalar', size=12)
plt.ylabel('Total de Crianças Admitidas', size=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### *Análise do Gráfico*

- A ala de Oncologia e a UTI Pediátrica são as que concentram o maior número de crianças admitidas.

- Isso sugere que os esforços para otimização de cardápio e redução de desperdício podem ter um impacto maior se focados nessas duas alas.

Vamos analisar também a distribuição por faixa etária, agora usando um gráfico de barras para uma comparação mais clara.

```python
# Gráfico de barras para o total de crianças admitidas por faixa etária
plt.figure(figsize=(10, 6))
sns.barplot(data=admissions_daily_df, x='faixa_etaria', y='total_de_crianças_admitidas', estimator=sum, errorbar=None, palette='magma')

plt.title('Total de Crianças Admitidas por Faixa Etária', size=16, fontweight='bold')
plt.xlabel('Faixa Etária', size=12)
plt.ylabel('Total de Crianças Admitidas', size=12)
plt.tight_layout()
plt.show()
```

### *Análise do Gráfico*

A distribuição de admissões por faixa etária segue o mesmo padrão da de rejeições: Crianças são o grupo mais numeroso, seguidas por Adolescentes e, por último, Bebês.

A alta quantidade de crianças admitidas reforça a importância de investigar a fundo as causas da rejeição de refeições nesse grupo específico.

### 4. Análise de Refeições Rejeitadas

Qual prato é o "vilão" da cozinha? Aqui, identificamos os pratos com maior número de rejeições. Essa informação é crucial para a equipe de nutrição, que pode optar por substituir ou adaptar as receitas menos aceitas.

```python
# Gráfico de barras para os 5 pratos mais rejeitados
top_5_rejected = rejected_meals_df.sort_values('rejected', ascending=False).head(5)

plt.figure(figsize=(12, 7))
sns.barplot(data=top_5_rejected, x='rejected', y='menu_name', palette='rocket')

plt.title('Top 5 Pratos Mais Rejeitados', size=16, fontweight='bold')
plt.xlabel('Quantidade de Rejeições', size=12)
plt.ylabel('Prato', size=12)
plt.tight_layout()
plt.show()
```

### *Análise do Gráfico*

Pratos como Sopa de Legumes, Fígado e Peixe cozído  estão entre os mais rejeitados.

Normalmente, são pratos geralmente rejeitados entre crianças. A alta rejeição pode estar relacionada a fatores como a gosto, forma de preparo, tempero, aparência ou a condição clínica das crianças, que pode afetar o apetite. Uma análise qualitativa paralela a uma substituição dos pratos seria um bom próximo passo.

### 5. Análise de Consumo e Desperdício

O desperdício é um ponto crítico em qualquer serviço de alimentação. Vamos analisar a taxa de desperdício (waste_ratio), que é a proporção de comida servida que não foi consumida. Faremos essa análise por ala e por tipo de refeição.

```python
# Gráfico de barras agrupado para taxa de desperdício por ala e tipo de refeição
plt.figure(figsize=(14, 8))
sns.barplot(data=ward_consumption_df, x='ward', y='waste_ratio', hue='meal_type', palette='crest')

plt.title('Taxa Média de Desperdício por Ala e Tipo de Refeição', size=16, fontweight='bold')
plt.xlabel('Ala Hospitalar', size=12)
plt.ylabel('Taxa de Desperdício Média (%)', size=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Tipo de Refeição')
plt.tight_layout()
plt.show()
```

### *Análise do Gráfico*

A UTI Pediátrica e a Cardiologia Pediátrica apresentam as maiores taxas de desperdício, especialmente no Café da Manhã.

Pacientes em UTIs frequentemente têm restrições alimentares ou dificuldades de alimentação, o que pode explicar a alta taxa de desperdício.

O café da manhã parece ser um ponto de atenção em quase todas as alas, com taxas de desperdício consistentemente mais altas em comparação com o almoço e o jantar.

### 6. Conclusão

Após a análise exploratória dos dados, podemos extrair os seguintes insights principais:

- Foco na Faixa Etária "Criança": Tanto em número de admissões quanto em rejeições, as crianças (não bebês, nem adolescentes) são o grupo mais representativo. Ações direcionadas a este público, como a criação de cardápios mais atrativos ou lúdicos, podem gerar o maior impacto na redução da rejeição.

- Pratos a Serem Revistos: O top 5 de pratos rejeitados (Macarrão, Panqueca, etc.) deve ser analisado pela equipe de nutrição. Testar variações nas receitas ou substituições pode diminuir a rejeição.

- Atenção às Alas de Oncologia e UTI: Sendo as alas com maior população, a otimização de processos e cardápios nelas pode trazer os melhores resultados em escala. A alta taxa de desperdício na UTI, em particular, pode indicar a necessidade de porções menores ou de uma comunicação mais próxima com a equipe de enfermagem sobre a condição dos pacientes.

- Oportunidade no Café da Manhã: O café da manhã é consistentemente a refeição com maior taxa de desperdício em todas as alas. Investigar as causas (horário, tipo de alimento, etc.) pode levar a uma redução significativa do desperdício geral.

Este notebook fornece um ponto de partida sólido para investigações mais aprofundadas e para a tomada de decisões baseada em dados na gestão da cozinha hospitalar.
